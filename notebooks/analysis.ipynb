```
import os, sys, json
import pandas as pd
sys.path.insert(0, os.path.abspath('..'))
```

## Load JSON Result Files

```
results_dir = "../results"
files = [f for f in os.listdir(results_dir) if f.startswith("results_")]
files
```

```
rows = []

for f in files:
    data = json.load(open(os.path.join(results_dir, f)))
    for entry in data['leaderboard']:
        rows.append({
            "file": f,
            "name": entry['name'],
            "role": entry['role'],
            "score": entry['score'],
            "tests": entry['tests']
        })

df = pd.DataFrame(rows)
df
```

```
df.groupby('name')['score'].mean().sort_values(ascending=False)
```
